```python
import pandas as pd

# load dataset
df = pd.read_csv('FitAnalytics.csv')

# data visualization
df.head()
```

```python
# columns informations

df.info()
```

```python
# columns describe

df.describe()
```

```python
# count workout categories

df['Categoria_Treino'].value_counts()
```

```python
# count Major Muscle Group

df['Grupo_Muscular_Principal'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Workout Category Bar Chart

df['Categoria_Treino'].value_counts().plot(kind='bar')
plt.title('Distribuição de Tipos de Treino')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create crosstab
heatmap_data = pd.crosstab(df['Objetivo_Treino'], df['Grupo_Muscular_Principal'])

# plot
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt='d')
plt.title('Objetivo do Treino vs Grupo Muscular')
plt.show()
```
